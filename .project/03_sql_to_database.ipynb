# Template for task: Create Database Tables From The GenAI Response

```python
import duckdb
import os
```

```python
def create_table_from_sql_files(con, sql_table_folder):
    sql_files = sorted([file for file in os.listdir(sql_table_folder) if file.endswith('.sql')])
    for file in sql_files:
        sql_file_path = os.path.join(sql_table_folder, file)
        with open(sql_file_path, 'r') as f:
            sql_script = f.read()
 
            try:
                con.execute(sql_script)
            except Exception as e:
                print(f"Error executing script: {sql_file_path}\nException: {e}")
                continue
 
            print(f"SQL table {sql_file_path} created")
            print("\n-----------------\n")
```

```python
# Connect to duckdb database and create tables from SQL files
con = duckdb.connect('alpha-app.duckdb')
 
create_table_from_sql_files(con, 'genai_design/table')
 
con.close()
```
